```python
# Visualization
# Distribution of trip miles
sns.histplot(data['distance'], bins=50)
plt.title('Distribution of Distance')
plt.show()
```

```python
# Plot the heatmap of correlations
plt.figure(figsize=(15, 6))
sns.heatmap(data1.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Feature Correlations')
plt.show()
```

```python
# Convert pickup_datetime to datetime format
data['pickup_datetime'] = pd.to_datetime(data['pickup_datetime'])

# Create a new column for the trip date
data['Trip Date'] = data['pickup_datetime'].dt.date

# Group by Trip Date and count the number of trips per day
trips_per_day = data.groupby('Trip Date').size()

# Reset the index if you want a DataFrame output
trips_per_day = trips_per_day.reset_index(name='Number of Trips')

print(trips_per_day)
```

```python
data['Taxi speed'] = data['distance'] * 1000 / data['Trip Seconds']
```

```python
data
```

```python
# Plot the number of trips per day
trips_per_day.plot(kind='line', marker='o', title='Number of Trips per Day')
plt.xlabel('Date')
plt.ylabel('Number of Trips')
plt.grid(True)
plt.show()
```

```python
# Group by hour and count the number of trips
trips_per_hour = data.groupby('hour').size()

# Plot the number of trips per hour
trips_per_hour.plot(kind='bar', title='Number of Trips per Hour')
plt.xlabel('Hour')
plt.ylabel('Number of Trips')
plt.grid(True)
plt.show()
```

```python
# Group by day of week and count the number of trips
trips_per_day_of_week = data.groupby('day_of_week').size()

# Plot the number of trips per day of week
trips_per_day_of_week.plot(kind='bar', title='Number of Trips per Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Trips')
plt.xticks(rotation=0)  # Keep x labels readable
plt.grid(True)
plt.show()
```

Speed vs Temperature

```python
plt.scatter(data['tempearture'], data['Taxi speed'], alpha=0.5)
plt.title('Speed vs Temperature')
plt.xlabel('Temperature (F)')
plt.ylabel('Speed (mph)')
plt.grid(True)
plt.show()
```

Speed vs Precipitation


```python
plt.scatter(data['precipitation'], data['Taxi speed'], alpha=0.5)
plt.title('Speed vs Precipitation')
plt.xlabel('Precipitation (inches)')
plt.ylabel('Speed (mph)')
plt.grid(True)
plt.show()
```

Speed vs Humidity

```python
plt.scatter(data['humidity'], data['Taxi speed'], alpha=0.5)
plt.title('Speed vs Humidity')
plt.xlabel('Humidity (%)')
plt.ylabel('Speed (mph)')
plt.grid(True)
plt.show()
```

Speed vs Wind Speed

```python
plt.scatter(data['wind_speed'], data['Taxi speed'], alpha=0.5)
plt.title('Speed vs Wind Speed')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Speed (mph)')
plt.grid(True)
plt.show()
```

Speed vs Rush Hour

```python
plt.boxplot([data[data['is_rush_hour'] == 0]['Taxi speed'], data[data['is_rush_hour'] == 1]['Taxi speed']], labels=['Non-Rush Hour', 'Rush Hour'])
plt.title('Speed vs Rush Hour')
plt.ylabel('Speed (mph)')
plt.grid(True)
plt.show()
```
